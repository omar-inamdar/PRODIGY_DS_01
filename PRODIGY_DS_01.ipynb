IMPORTING THE LIBRARIES



```python
import pandas as pd
import matplotlib.pyplot as plt
```

READING DATASET

```python
data = pd.read_csv('HRDataset_v13.csv')
```

```python
data.head()
```

```python
data.info()
```

PLOTTING THE HISTOGRAM TO SHOW REPRESENTATION

```python
gender_employment_source_counts = data.groupby(['Sex', 'EmploymentStatus', 'RecruitmentSource']).size().unstack()


fig, ax = plt.subplots(figsize=(10, 5))
gender_employment_source_counts.plot(kind='hist', stacked=True, ax=ax)

ax.set_xlabel('Gender and Employment Status')
ax.set_ylabel('Count')
ax.set_title('Distribution of Employment Status and Recruitment Source by Gender')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
```
